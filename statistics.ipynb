```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df_btc = pd.read_csv('./datasets/btc_all_time.csv')
df_eth = pd.read_csv('./datasets/eth_all_time.csv')
df_xrp = pd.read_csv('./datasets/xrp_all_time.csv')
df_doge = pd.read_csv('./datasets/doge_all_time.csv')
df_aave = pd.read_csv('./datasets/aave_all_time.csv')
```

```python
for df in [df_btc, df_eth, df_xrp, df_doge, df_aave]:
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
    scaler = MinMaxScaler()
    df['Scaled_Close'] = scaler.fit_transform(df['Close'].values.reshape(-1, 1))
    df['Scaled_High'] = scaler.fit_transform(df['High'].values.reshape(-1, 1))
    df['Scaled_Low'] = scaler.fit_transform(df['Low'].values.reshape(-1, 1))
    df['Scaled_Open'] = scaler.fit_transform(df['Open'].values.reshape(-1, 1))
    df['Scaled_Volume'] = scaler.fit_transform(df['Volume'].values.reshape(-1, 1))
    df['Scaled_Market_Cap'] = scaler.fit_transform(df['Market Cap'].values.reshape(-1, 1))
```

```python
df = df_btc.drop(columns=['Close', 'High', 'Low', 'Open', 'Volume', 'Market Cap', 'Time Open', 'Time Close', 'Time High', 'Time Low'])
df.head()
```

```python
# Plot of Close price for all coins
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_btc, x='Date', y='Scaled_Close', marker='o')
sns.lineplot(data=df_eth, x='Date', y='Scaled_Close', marker='o')
sns.lineplot(data=df_xrp, x='Date', y='Scaled_Close', marker='o')
sns.lineplot(data=df_doge, x='Date', y='Scaled_Close', marker='o')
sns.lineplot(data=df_aave, x='Date', y='Scaled_Close', marker='o')
plt.title('Close Price of Cryptocurrencies')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)

plt.legend()
plt.show()
```
